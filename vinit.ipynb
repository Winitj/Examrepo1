```python
import dask.dataframe as dd
from sqlalchemy import create_engine
import os

# Define your MySQL connection string
mysql_conn_str = 'mysql+mysqlconnector://root:Vinit2003@localhost:3306/vinit'

# Create a SQLAlchemy engine
engine = create_engine(mysql_conn_str)

# Specify the MySQL table name
table_name = 'up'

# Path to the directory containing CSV files
csv_directory = 'C://path/to/your/csv/files/'

# Get a list of CSV files in the directory
csv_files = [file for file in os.listdir(csv_directory) if file.endswith('.csv')]

# Loop through each CSV file
for csv_file in csv_files:
    # Create a Dask DataFrame from CSV
    ddf = dd.read_csv(os.path.join(csv_directory, csv_file))

    # Convert Dask DataFrame to Pandas DataFrame
    pdf = ddf.compute()

    # Use the `to_sql` method to write Pandas DataFrame to MySQL
    pdf.to_sql(name=table_name, con=engine, if_exists='append', index=False, method='multi')

print("Data uploaded successfully.")
```
